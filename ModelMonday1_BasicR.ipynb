## Basic R Commands

```r
# Creating a vector

x1 <- c(1, 3, 2, 5)
x2 = c(3, 2, 4, 5)
```

```r
print(x1)
print(x2)
```

```r
# Adding vectors together, piecewise

x4 = x1 + x2
```

```r
# Using ls() 

ls()
```

```r
# Using rm()

rm(x4)
```

```r
# Use ? to find out what functions do

?ls()
```

```r
# Create a matrix

hundo_matrix = matrix(c(1:100), 10, 10, byrow = F)
hundo_matrix
```

```r
# Perform matrix operations
hundo_matrix + hundo_matrix
```

## Graphing

```r
# Plot a basic plot
x = rnorm(100)
y = x + rnorm(100, mean=10, sd=1)

plot(x, y, xlab = 'Yoga Classes', ylab = 'Beets Eaten', main = 'Beet Yoga')
```

```r
# Questioning normal distribution - which is built right in!
?rnorm
```

```r
#

x = 1:100
y = 1:100
z = outer(x, y, function(a,b) (sin(a *b)))
```

```r
persp(x,y,z)
```
